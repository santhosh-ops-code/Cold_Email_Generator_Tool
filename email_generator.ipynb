```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://atliq.keka.com/careers/jobdetails/93894")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### INSTRUCTION:
    The scraped text below is from a careers page.

    Extract the job information and return ONLY valid JSON
    with the following keys:
    - role
    - experience
    - skills
    - description

    ### SCRAPED TEXT:
    {page_data}

    ### OUTPUT (JSON ONLY, NO EXTRA TEXT):
    """
)

```

```python
chain_extract = prompt_extract | llm

res = chain_extract.invoke(
    {"page_data": page_data}
)

print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser()
json_res  = json_parser.parse(res.content)
json_res
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient("vectorstore")
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(
            documents=row["Techstack"],
            metadatas={"links": row["Links"]},
            ids=[str(uuid.uuid4())]
        )
```

```python
from langchain_core.prompts import PromptTemplate

prompt_email = PromptTemplate.from_template(
    """
    ### JOB DESCRIPTION:
    {job_description}

    ### INSTRUCTION:
    You are Mohan, a Business Development Executive (BDE) at Atliq.
    Atliq is an AI & Software company that enables seamless integration
    of business processes through automated tools.

    Write a professional cold email to the client regarding the above job.
    Clearly explain how Atliq can help fulfill their requirements.

    ALWAYS start the email with:
    "Dear Hiring Manager,"

    Include the most relevant portfolio links from the list below.

    ### PORTFOLIO LINKS:
    {links}

    Rules:
    - Do NOT use placeholders like [Recipient Name]
    - Do NOT add a preamble or explanation
    - Output ONLY the email content

    ### EMAIL:
    """
)
```

```python
chain_email = prompt_email | llm

res = chain_email.invoke({
    "job_description": str(job),
    "links": links
})

print(res.content)
```
